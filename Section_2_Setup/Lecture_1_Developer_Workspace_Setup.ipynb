## üèóÔ∏è Step 5: Create Virtual Environment

Virtual environments isolate your project dependencies. Let's create one using the terminal.

### üîß Terminal Commands:

**If UV is working (from previous step):**
```bash
uv venv .venv
```

**If UV failed, use standard Python:**
```bash
python -m venv .venv
```

### üéØ Activate Virtual Environment:

**Windows:**
```bash
.venv\Scripts\activate
```

**Mac/Linux:**
```bash
source .venv/bin/activate
```

**‚úÖ Success Indicator**: Your terminal prompt should show `(.venv)` at the beginning

**üí° VS Code Tip**: VS Code will automatically detect and use this environment

**After running the commands above, run the verification cell below:**

````python
# Run this cell AFTER creating your virtual environment
from pathlib import Path
import sys

print("üîç Virtual Environment Check")
print("="*30)

venv_path = Path(".venv")
if venv_path.exists():
    print("‚úÖ .venv folder exists!")
    
    # Check for Python executable
    if sys.platform == "win32":
        python_exe = venv_path / "Scripts" / "python.exe"
        activate_cmd = ".venv\\Scripts\\activate"
    else:
        python_exe = venv_path / "bin" / "python"
        activate_cmd = "source .venv/bin/activate"
    
    if python_exe.exists():
        print(f"‚úÖ Python executable found!")
        print(f"\nüí° To activate: {activate_cmd}")
        print("üí° Look for (.venv) in your terminal prompt!")
    else:
        print("‚ùå Python executable missing")
else:
    print("‚ùå .venv folder not found")
    print("Please run the terminal commands from Step 5")

print("\nüéØ Virtual environment setup complete!")
````

**After running the installation commands, run the verification cell below:**

````python
# Run this cell AFTER installing packages
print("üß™ Testing Package Installation...")
print("="*35)

# Test basic Python packages (should always work)
basic_packages = ["os", "sys", "pathlib", "json"]
print("Basic Python packages:")
for pkg in basic_packages:
    try:
        __import__(pkg)
        print(f"  ‚úÖ {pkg}")
    except:
        print(f"  ‚ùå {pkg}")

# Test AI packages (from requirements.txt)
print("\nAI packages (from requirements.txt):")
ai_packages = ["openai", "jupyter"]

installed_count = 0
for pkg in ai_packages:
    try:
        __import__(pkg)
        print(f"  ‚úÖ {pkg} - Installed successfully!")
        installed_count += 1
    except ImportError:
        print(f"  ‚ùå {pkg} - Not installed")
    except Exception as e:
        print(f"  ‚ö†Ô∏è  {pkg} - Issue: {str(e)[:30]}...")

print(f"\nüìä AI Packages: {installed_count}/{len(ai_packages)} installed")

if installed_count == len(ai_packages):
    print("üéâ Perfect! All packages installed successfully!")
elif installed_count > 0:
    print("‚úÖ Most packages working - you're ready to continue!")
else:
    print("‚ùå Packages not installed - please run the terminal commands from Step 6")

print("\nüöÄ Package installation check complete!")
````

## üÜò Quick Troubleshooting

**üêç Python Issues:**
- **Not found?** Download from python.org and check "Add to PATH"
- **Wrong version?** Use Python 3.9-3.11 for best compatibility

**üì¶ Package Issues:**
- **UV installation fails?** Use pip instead - works perfectly!
- **Package conflicts?** Delete .venv folder and recreate:
  ```bash
  # Delete .venv folder manually in VS Code
  python -m venv .venv
  # Activate and reinstall packages
  ```

**üíª VS Code Issues:**
- **Python not detected?** Press `Ctrl+Shift+P` ‚Üí "Python: Select Interpreter"
- **Choose interpreter in .venv folder**
- **Jupyter not working?** Install Jupyter extension and restart VS Code

**üîí Environment Variables:**
- **File not working?** Ensure it's named exactly `.env`
- **Format:** `KEY=value` (no spaces around =)
- **Restart VS Code** after creating .env file
